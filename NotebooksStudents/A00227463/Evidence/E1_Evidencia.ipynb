## A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
from sklearn.cluster import KMeans

k = 5
kmeans_model = KMeans(n_clusters=k, random_state=1).fit(df)
```

Add to your dataset a column with the estimated cluster to each data point

```python
df['cluster'] = kmeans_model.labels_

sns.scatterplot(x=df['x1'], y=df['x2'], hue=df['cluster'], palette='viridis')
```

Print the number associated to each cluster

```python
# number associated with each cluster
print(kmeans_model.labels_)
```

Print the centroids

```python
# print centroids 
print(kmeans_model.cluster_centers_)
```

Print the intertia metric

```python
# print inertia
print(kmeans_model.inertia_)
```

## A5 Elbow plot

Compute the Elbow plot

```python
# compute elbow plot
inertia = []
k_values = range(1, 20)
for k in k_values:
    kmeans_model = KMeans(n_clusters=k, random_state=1).fit(df)
    inertia.append(kmeans_model.inertia_)
sns.lineplot(x=k_values, y=inertia)
```
