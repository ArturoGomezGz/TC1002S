## A4 Kmeans

Do Kmeans clustering assuming a number of clusters according to your scatter plot

```python
from sklearn.cluster import KMeans
import numpy as np

# Combining x1 and x2 into a single array for clustering
data = np.column_stack((variable_1, variable_2))

# Assuming 3 clusters based on visual inspection of the scatter plot
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)

# Adding cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Plotting the scatter plot with cluster assignments
plt.figure(figsize=(10, 6))
plt.scatter(variable_1, variable_2, c=kmeans.labels_, cmap='viridis', alpha=0.7)
plt.title('K-Means Clustering of x1 vs. x2')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.colorbar(label='Cluster')
plt.show()
```

## A5 Elbow plot

Compute the Elbow plot

```python
# Data preparation
data = np.column_stack((variable_1, variable_2))

# Calculating inertia for different numbers of clusters
inertia_values = []
cluster_range = range(1, 11)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia_values.append(kmeans.inertia_)

# Plotting the Elbow plot
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='-')
plt.title('Elbow Plot for KMeans Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.xticks(cluster_range)
plt.show()
```

## Questions

**What is the best number of clusters K? (argue your response)**

The best number of clusters, K, appears to be **3** based on the Elbow plot. This is because the plot shows a noticeable "elbow" or bend at K = 3, indicating a significant reduction in inertia up to this point. Beyond K = 3, the rate of decrease in inertia slows down, suggesting diminishing returns from adding more clusters. The elbow point signifies the balance between having enough clusters to capture the data's variability without overfitting or adding unnecessary complexity.

**Does this number of clusters agree with your initial guess? (argue your response, no problem at all if they do not agree)**

Yes, this number of clusters agrees with the initial guess of 3 clusters made by observing the scatter plot. The scatter plot visually suggested three distinct groupings, and the KMeans clustering confirmed this by placing centroids that aligned well with the data's natural separations. Both the initial visual guess and the Elbow plot indicate that 3 clusters capture the underlying structure of the data effectively, validating the initial assessment made before clustering.
