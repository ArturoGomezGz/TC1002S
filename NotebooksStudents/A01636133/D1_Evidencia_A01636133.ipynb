## A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
from sklearn.cluster import KMeans

# Define number of clusters
K  = 2

# Create/Initialize the Kmeans box/object
km = KMeans(n_clusters=K, n_init="auto")

# Do K-means clustering (assing each point in the dataset to a cluster)
newdf = df.drop('Unnamed: 0', axis=1)
yestimated = km.fit_predict(newdf)

# Print estimated cluster of each point in the dataset
yestimated
```

Add to your dataset a column with the estimated cluster to each data point

```python
newdf['yestimated'] = yestimated

newdf
```

Print the number associated to each cluster

```python
# Print the Labels/Names of the existing clusters
newdf.yestimated.unique()
```

Print the centroids

```python
km.cluster_centers_
```

Print the intertia metric

```python
km.inertia_
```

```python
newdf
```

Plot a scatter plot of your data using different color for each cluster. Also plot the centroids

```python
df1 = newdf[newdf.yestimated == 0]
df2 = newdf[newdf.yestimated == 1]

plt.scatter(df1.x1, df1.x2, label='Cluster 0', c='r', marker='o', s=32, alpha=0.3)
plt.scatter(df2.x1, df2.x2, label='Cluster 1', c='b', marker='s', s=32, alpha=0.3)

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='black', marker='*', label='Centroides', s=256)

plt.title('Scatter plot (for each cluster)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

## A5 Elbow plot

Compute the Elbow plot

```python
# Intialize a list to hold sum of squared error (sse)
sse = []

# Define values of k
k_rng = range(1,10)

# For each k
for k in k_rng:
    # Create model
    km = KMeans(n_clusters=k, n_init="auto")
    # Do K-means clustering
    km.fit_predict(newdf[['x1','x2']])
    # Save sse for each k
    sse.append(km.inertia_)
```

```python
# Plot sse versus k
plt.plot(k_rng,sse, 'o-', markersize=8)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```
