## d) Elbow plot

Compute the Elbow plot

```python
inertia_values = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(dataset.iloc[:, 1:])  
    inertia_values.append(kmeans.inertia_)

# Plot the Elbow plot
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia_values, 'bo-')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Elbow graph')
plt.show()
```

3) Do K means clustering in the following cases:

* KmeansAll: Using all 64 variables/pixels/features

* Kmeans1row: Using only the 8 variables/pixels/features from the firt row

* Kmeans4row: Using only the 8 variables/pixels/features from the fourth row

* Kmeans8row: Using only the 8 variables/pixels/
features from the eighth row


```python
# KmeansAll: Usando los 64  pixeles
#usaré k=10 ya que hay 10 dígitos posibles
kmeans_all = KMeans(n_clusters=10, random_state=42)
kmeans_all.fit(digits.data)
digits['ClusterTodo'] = kmeans_all.labels_
```

```python

# Kmeans1row
kmeans_1row = KMeans(n_clusters=10, random_state=42)
#todas las columnas de la primera fila, cada fila es una imagen de 8x8 pixeles, por lo que la primera fila son los primeros 8 pixeles
kmeans_1row.fit(digits.data[:, :8])
digits['Cluster1Fila'] = kmeans_1row.labels_
```

```python

# Kmeans4row
kmeans_4row = KMeans(n_clusters=10, random_state=42)
kmeans_4row.fit(digits.data[:, 24:32])
digits['Cluster4Fila'] = kmeans_4row.labels_
```

```python

# Kmeans8row
kmeans_8row = KMeans(n_clusters=10, random_state=42)
kmeans_8row.fit(digits.data[:, 56:])
digits['Cluster8Fila'] = kmeans_8row.labels_
```

```python

# Imprimir el número de muestras en cada cluster para cada caso
print("Conteo de clusters en KmeansTodo:", np.bincount(digits['ClusterTodo']))
print("Conteo de clusters en Kmeans1Fila:", np.bincount(digits['Cluster1Fila']))
print("Conteo de clusters en Kmeans4Fila:", np.bincount(digits['Cluster4Fila']))
print("Conteo de clusters en Kmeans8Fila:", np.bincount(digits['Cluster8Fila']))
```

4) Verify your results. Plot several observations from the same digit and add in the title the real label and the estimated label to check in what observations the clusterization was correct or incorrect

```python
num_samples = 5  # para cada número del 0 al 9, mostraré 5 ejemplos
fig, axes = plt.subplots(10, num_samples, figsize=(10, 10))

for digit in range(10):
    #obtengo los índices de las imágenes que corresponden al dígito actual
    digit_indices = np.where(digits.target == digit)[0]
    #selecciono num_samples índices al azar
    selected_indices = np.random.choice(digit_indices, num_samples, replace=False)
    
    #para cada índice seleccionado, mostraré la imagen y el cluster asignado
    for i, index in enumerate(selected_indices):
        ax = axes[digit, i]
        ax.imshow(digits.images[index], cmap=plt.cm.gray_r)
        #aqui se cambia el cluster a comparar con el valor real
        ax.set_title(f'True: {digits.target[index]}\nPred: {digits["ClusterTodo"][index]}')
        ax.axis('off')
plt.tight_layout()
```

5) Compute the Elbow plot

```python
inercia = []
valores_k = range(1, 15)

for k in valores_k:
    kmeans_new = KMeans(n_clusters=k, random_state=42)
    kmeans_new.fit(digits.data)  
    inercia.append(kmeans_new.inertia_)

# Plot the new Elbow plot
plt.figure(figsize=(10, 6))
plt.plot(valores_k, inercia, 'bo-')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Nuevo Elbow graph')
plt.show()
```
