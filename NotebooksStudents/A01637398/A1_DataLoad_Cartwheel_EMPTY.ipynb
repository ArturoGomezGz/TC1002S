Iris

```python
# Importar librer√≠as necesarias
import pandas as pd
from sklearn.datasets import load_iris

# Cargar el dataset de iris
iris = load_iris()

# Crear un DataFrame con las caracter√≠sticas de iris
iris_df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])

# A√±adir la columna de especies (target)
iris_df['species'] = iris['target']

# Mostrar las primeras filas del DataFrame
print(iris_df.head())

# Explorar el DataFrame
print(f"Shape of the iris dataset: {iris_df.shape}")
print(f"Column names: {iris_df.columns}")
print(f"Descriptive statistics: \n{iris_df.describe()}")

# Estad√≠sticas por especie
species_mapping = dict(enumerate(iris['target_names']))
print(f"Species mapping: {species_mapping}")

# Reemplazar valores num√©ricos de 'species' con nombres
iris_df['species'] = iris_df['species'].map(species_mapping)

# Mostrar DataFrame con nombres de especies
print(iris_df.head())
```
