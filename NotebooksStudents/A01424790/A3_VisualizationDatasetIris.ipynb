```python
url = "/Users/fernando/Pictures/Escul/semanaTec_1/TC1002S/NotebooksProfessor/datasets/iris/iris.csv"

dataf = pd.read_csv(url)
dataf.columns = ["LargoSepalo", "AnchoSepalo", "LargoPetalo", "AnchoSepalo", "Tipo_flor"]
```

```python
#2. Plot the histograms for each of the quantitative variables
sns.displot(dataf["LargoSepalo"], kde = False)
dfhist = sns.displot(dataf["AnchoSepalo"], kde = False)
dfhist_2 = sns.displot(dataf["LargoPetalo"], kde = False)
dfhist_3 = sns.displot(dataf["AnchoSepalo"], kde = False)

plt.show()
```

```python
#3 Plot the boxplots for each of the quantitative variables
x = dataf.loc[:, ["LargoSepalo", "AnchoSepalo", "LargoPetalo", "AnchoSepalo", "Tipo_flor"]]
x2bp = sns.boxplot(data=x, orient="V")
x2bp.set_title("Box plot of the Height and width of the petal and setal")
plt.ylabel("cm")
plt.show()
```
