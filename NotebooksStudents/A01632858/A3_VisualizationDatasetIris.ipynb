Exploring the content of the data set

```python
df = pd.read_csv('myNewData.csv')
df.rename(columns={"4":"Type of flower"}, inplace=True)
df.rename(columns={"0":"Sepal length"}, inplace=True)
df.rename(columns={"1":"Sepal width"}, inplace=True)
df.rename(columns={"2":"Petal length"}, inplace=True)
df.rename(columns={"3":"Petal width"}, inplace=True)
df.drop(columns=["Unnamed: 0"], inplace=True)
df.head()
```

Frequency_tables

```python
#Number of times that each distinct value of a variable occurs in data set
df['Type of flower'].value_counts()
```

```python
#Proportion of each distinct value of a variable occurs in data set
x = df['Type of flower'].value_counts()
x/x.sum()
```

Histogram

```python
sns.displot(df, kde = False)
plt.title("Histogram of quantitative values")
plt.ylabel("Count")
plt.show()
```

```python
#Plot histogram of the total bill only
sns.displot(df['Sepal length'], kde = False)
plt.ylabel("Count")
plt.title("Quantitative Sepal length")
plt.show()
```

```python
#Plot distribution of the tips only
sns.displot(df['Sepal width'], kde = True)
plt.ylabel("Count")
plt.title("Quantitative Sepal width")
plt.show()
```

```python
#Plot histogram of both the age and the wingspan
sns.displot(df['Petal length'], kde = True)
plt.ylabel('Count')
plt.title("Quantitaive petal length")
plt.show()
```

```python
sns.displot(df['Petal width'], kde = True)
plt.ylabel('Count')
plt.title("Quantative petal width")
plt.show()
```

```python
sns.displot(df['Type of flower'], kde = True)
plt.ylabel('Count')
plt.title("Quantative types of flower")
plt.show()
```

```python
sns.boxplot(df)
plt.title('Flowers stadistics boxplot')
plt.show()
```

```python
x = df.loc[:, ['Petal width', 'Type of flower']]
x2bp = sns.boxplot(data=x, orient="v")
plt.ylabel('Centimeters')
plt.show()
```

```python
y = df.loc[:, ['Sepal length', 'Type of flower']] 
y2bp = sns.boxplot(data=y, orient="v")
plt.ylabel('Centimeters')
plt.show()
```
