## d) Elbow plot

Compute the Elbow plot

```python

inertia_values = []
cluster_range = range(1, 11)  # Probando de 1 a 10 clusters

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia_values.append(kmeans.inertia_)


plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='--')
plt.title('Elbow Plot para determinar el número óptimo de clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()
```

3) Do K means clustering in the following cases:

* KmeansAll: Using all 64 variables/pixels/features

* Kmeans1row: Using only the 8 variables/pixels/features from the firt row

* Kmeans4row: Using only the 8 variables/pixels/features from the fourth row

* Kmeans8row: Using only the 8 variables/pixels/
features from the eighth row


```python
from sklearn.cluster import KMeans
import numpy as np

# Cargar el dataset de dígitos
digits = load_digits()

# Caso 1: KmeansAll - Usando todas las 64 variables/píxeles/características
X_all = digits.data  # Usa todas las 64 características
kmeans_all = KMeans(n_clusters=10, random_state=42)  # Asumimos 10 clusters (correspondientes a los dígitos 0-9)
kmeans_all.fit(X_all)

# Caso 2: Kmeans1row - Usando solo las 8 variables de la primera fila
X_1row = digits.data[:, :8]  # Usando las primeras 8 características
kmeans_1row = KMeans(n_clusters=10, random_state=42)
kmeans_1row.fit(X_1row)

# Caso 3: Kmeans4row - Usando solo las 8 variables de la cuarta fila
X_4row = digits.data[:, 24:32]  # Usando las características de la cuarta fila (índices 24 a 31)
kmeans_4row = KMeans(n_clusters=10, random_state=42)
kmeans_4row.fit(X_4row)

# Caso 4: Kmeans8row - Usando solo las 8 variables de la octava fila
X_8row = digits.data[:, 56:64]  # Usando las características de la octava fila (índices 56 a 63)
kmeans_8row = KMeans(n_clusters=10, random_state=42)
kmeans_8row.fit(X_8row)

# Imprimir los resultados de inercia para cada caso
print(f"Inercia KmeansAll (todas las características): {kmeans_all.inertia_}")
print(f"Inercia Kmeans1row (primer fila de características): {kmeans_1row.inertia_}")
print(f"Inercia Kmeans4row (cuarta fila de características): {kmeans_4row.inertia_}")
print(f"Inercia Kmeans8row (octava fila de características): {kmeans_8row.inertia_}")
```

4) Verify your results. Plot several observations from the same digit and add in the title the real label and the estimated label to check in what observations the clusterization was correct or incorrect

```python
import matplotlib.pyplot as plt
import numpy as np

# Número de observaciones por dígito a visualizar
num_samples_per_digit = 3  # Puedes cambiar este número según cuántas observaciones quieras ver por dígito

# Crear el gráfico para verificar los resultados
plt.figure(figsize=(12, 15))

# Iterar sobre cada dígito (0 a 9)
for digit_to_check in range(10):
    # Filtrar las observaciones del dígito actual
    indices = np.where(digits.target == digit_to_check)[0]

    for i in range(num_samples_per_digit):
        # Seleccionar la imagen
        idx = indices[i]

        # Etiqueta real y etiqueta estimada por KMeans
        true_label = digits.target[idx]
        estimated_label = kmeans_all.labels_[idx]  # Cambiar si quieres verificar kmeans_1row, kmeans_4row, kmeans_8row

        # Graficar la imagen
        plt.subplot(10, num_samples_per_digit, digit_to_check * num_samples_per_digit + i + 1)
        plt.imshow(digits.images[idx], cmap='gray')
        plt.title(f'Real: {true_label}\nEstimado: {estimated_label}')
        plt.axis('off')  # Quitar los ejes para mayor claridad

plt.suptitle('Comparación de etiquetas reales vs estimadas por K-Means', y=1.02)
plt.tight_layout()
plt.show()
```

5) Compute the Elbow plot

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Usar todas las características (64 píxeles)
X_all = digits.data

# Rango de número de clusters a probar
cluster_range = range(1, 11)

# Lista para almacenar los valores de inercia para cada número de clusters
inertia_values = []

# Calcular KMeans para diferentes números de clusters
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_all)
    inertia_values.append(kmeans.inertia_)

# Graficar el Elbow plot
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='--')
plt.title('Elbow Plot para determinar el número óptimo de clusters (KMeansAll)')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.grid(True)
plt.show()
```

4) Tune a classifier (Use the train set) in the following cases:

* ClassifierAll: Using all 64 variables/pixels/features

* Classifier1col: Using only the 8 variables/pixels/features from the firt column

* Classifier4col: Using only the 8 variables/pixels/features from the fourth column

* Classifier8col: Using only the 8 variables/pixels/
features from the eighth column

Note: in these four cases always use the same classification algorithm, e.g., a Suport Vector Machine


```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Cargar el dataset de dígitos
digits = load_digits()

# Dividir el dataset en train y test
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)

# 1. ClassifierAll: Usando todas las características (64 píxeles)
clf_all = SVC(random_state=42)
clf_all.fit(X_train, y_train)
y_pred_all = clf_all.predict(X_test)
accuracy_all = accuracy_score(y_test, y_pred_all)

# 2. Classifier1col: Usando solo la primera columna de píxeles (8 características)
X_train_1col = X_train[:, :8]  # Solo la primera columna de píxeles
X_test_1col = X_test[:, :8]
clf_1col = SVC(random_state=42)
clf_1col.fit(X_train_1col, y_train)
y_pred_1col = clf_1col.predict(X_test_1col)
accuracy_1col = accuracy_score(y_test, y_pred_1col)

# 3. Classifier4col: Usando solo la cuarta columna de píxeles (8 características)
X_train_4col = X_train[:, 24:32]  # Solo la cuarta columna de píxeles
X_test_4col = X_test[:, 24:32]
clf_4col = SVC(random_state=42)
clf_4col.fit(X_train_4col, y_train)
y_pred_4col = clf_4col.predict(X_test_4col)
accuracy_4col = accuracy_score(y_test, y_pred_4col)

# 4. Classifier8col: Usando solo la octava columna de píxeles (8 características)
X_train_8col = X_train[:, 56:64]  # Solo la octava columna de píxeles
X_test_8col = X_test[:, 56:64]
clf_8col = SVC(random_state=42)
clf_8col.fit(X_train_8col, y_train)
y_pred_8col = clf_8col.predict(X_test_8col)
accuracy_8col = accuracy_score(y_test, y_pred_8col)

# Imprimir las precisiones (accuracy) de cada clasificador
print(f"Accuracy ClassifierAll (todas las características): {accuracy_all}")
print(f"Accuracy Classifier1col (primera columna): {accuracy_1col}")
print(f"Accuracy Classifier4col (cuarta columna): {accuracy_4col}")
print(f"Accuracy Classifier8col (octava columna): {accuracy_8col}")
```
