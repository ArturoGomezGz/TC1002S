# Question 6


```python
total_variables = df.shape[1]

# Get the column names and their corresponding data types
column_info = df.dtypes

# Print the total number of variables and what they represent
print("Total number of variables (columns):", total_variables)
print("\nColumn names and their corresponding data types:")
for column_name, data_type in column_info.items():
    print(f"{column_name}: {data_type}")
```
