# A3 Activities

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Plot the histograms for each of the four quantitative variables

```python
quant_vars = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] 

# Plot histograms for each quantitative variable
plt.figure(figsize=(14, 8))  # Adjust the figure size as needed

for i, var in enumerate(quant_vars):
    plt.subplot(2, 2, i + 1)  # Adjust layout: 2x2 grid
    sns.histplot(df[var], bins=30, kde=True, color='skyblue')
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

2. Plot the histograms for each of the quantitative variables group by species

```python
quant_vars = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] 
groupvar = 'species'
plt.figure(figsize=(14, 10))  # Adjust the figure size as needed

for i, var in enumerate(quant_vars):
    plt.subplot(2, 2, i + 1)  # Adjust layout: 2x2 grid
    sns.histplot(data=df, x=var, hue=groupvar, multiple="stack", bins=30, palette='Set2')
    plt.title(f'Histogram of {var} grouped by {groupvar}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

3. Plot the boxplots for each of the quantitative variables

```python
quant_vars = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] 

# Plot histograms for each quantitative variable
plt.figure(figsize=(14, 8))  # Adjust the figure size as needed

for i, var in enumerate(quant_vars):
    plt.subplot(2, 2, i + 1)  # Adjust layout: 2x2 grid
    sns.boxplot(data=df, y=var, color='lightgreen')
    plt.title(f'Boxplot of {var}')
    plt.xlabel('')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
```

4. Plot the boxplots of the petal width grouped by type of flower

```python
plt.figure(figsize=(14, 8))
sns.boxplot(y='petal_width', hue='species', data=df, palette='Set2')
plt.title('Box Plot of petal width by species')
plt.ylabel('Petal width')
plt.xlabel('Species')
```

5. Plot the boxplots of the setal length grouped by type of flower

```python
plt.figure(figsize=(14, 8))

sns.boxplot(y='sepal_length', hue='species', data=df, palette='Set2')
plt.title('Box Plot of sepal length by species')
plt.ylabel('Sepal length')
plt.xlabel('Species')
```
