# Activity: work with the iris dataset

1. Load the iris.csv file in your computer and understand the dataset


2. How many observations (rows) are in total?


3. How many variables (columns) are in total? What do they represent?


4. How many observations are for each type of flower?


5. What is the type of data for each variable?


6. What are the units of each variable?


```python
# 1. Load and understand the dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
Ruta = "./iris/"
iris_df = pd.read_csv(Ruta + "iris.csv")

# Display first few rows to understand the structure
print("First 5 rows of the dataset:")
print(iris_df.head())
print("\nBasic information about the dataset:")
print(iris_df.info())

# 2. Count total observations (rows)
total_rows = len(iris_df)
print(f"\nTotal number of observations: {total_rows} rows")

# 3. Count and describe variables (columns)
total_cols = len(iris_df.columns)
print(f"\nTotal number of variables: {total_cols} columns")
print("\nVariables and what they represent:")
variables_description = {
    'sepal.length': 'Length of the sepal (the outer part of the flower)',
    'sepal.width': 'Width of the sepal',
    'petal.length': 'Length of the petal (the inner part of the flower)',
    'petal.width': 'Width of the petal',
    'variety': 'Type of iris flower'
}
for var, desc in variables_description.items():
    print(f"- {var}: {desc}")

# 4. Count observations per flower type
print("\nNumber of observations for each type of flower:")
print(iris_df['variety'].value_counts())

# 5. Data types for each variable
print("\nData type for each variable:")
print(iris_df.dtypes)

# 6. Units for each variable
print("\nUnits for each variable:")
units = {
    'sepal.length': 'Centimeters',
    'sepal.width': 'Centimeters',
    'petal.length': 'Centimeters',
    'petal.width': 'Centimeters',
    'variety': 'Categorical (no units)'
}
for var, unit in units.items():
    print(f"- {var}: {unit}")

# Bonus: Visual representation of the data
plt.figure(figsize=(15, 10))

# Create a pair plot to visualize relationships between variables
sns.pairplot(iris_df, hue='variety')
plt.suptitle('Pair Plot of Iris Dataset Variables by Variety', y=1.02)
plt.show()

# Summary statistics
print("\nSummary statistics for numerical variables:")
print(iris_df.describe())

# Adicional: Crear un gráfico de cajas para ver la distribución de las medidas por variedad
plt.figure(figsize=(15, 8))
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
measures = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

for i, measure in enumerate(measures):
    ax = axes[i//2, i%2]
    sns.boxplot(x='variety', y=measure, data=iris_df, ax=ax)
    ax.set_title(f'Distribution of {measure} by Variety')

plt.tight_layout()
plt.show()
```
