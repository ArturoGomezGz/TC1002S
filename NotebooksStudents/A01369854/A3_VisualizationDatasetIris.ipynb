# Frequency tables

The `value_counts()` method can be used to determine the number of times that each distinct value of a variable occurs in a data set.  In statistical terms, this is the "frequency distribution" of the variable. The `value_counts()` method produces a table with two columns.  The first column contains all distinct observed values for the variable.  The second column contains the number of times each of these values occurs.  Note that the table returned by `value_counts()` is actually a **Pandas** data frame, so can be further processed using any Pandas methods for working with data frames.


```python
# Number of times that each distinct value of a variable occurs in a data set
ms1 = df.value_counts()
ms1
```

```python
# Proportion of each distinct value of a variable occurs in a data set
propor = df.value_counts() / len(df)
propor
```

Note that the `value_counts()` method excludes missing values. We confirm this below by adding up observations to your data frame with some missing values and then computing `value_counts()` and comparing this to the total number of rows in the data set, which is 28. This tells us that there are 28 - (21+6) = 1 missing values for this variable (other variables may have different numbers of missing values).

```python
# Total number of observations

obser = len(df)
# Total number of null observations

nobser = df.isnull().sum()

# Total number of counts (excluding missing values)
tot = obser-nobser

tot
```

# Histogram

It is often good to get a feel for the shape of the distribution of the data.

```python
# Plot histogram of the Ms_1 only
plt.hist(df['Ms_1'], bins=len(df))  # Adjust the number of bins as needed
plt.title('Histogram of Ms_1')
plt.xlabel('Ms_1 Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of the Ms_2 only
plt.hist(df['Ms_2'], bins=len(df))  # Adjust the number of bins as needed
plt.title('Distribution of Ms_2')
plt.xlabel('Ms_2 Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot histogram of both the Ms_1 and the Ms_2
plt.hist(df['Ms_1'], bins=len(df), alpha=0.8, label='Ms_1')  # Adjust the number of bins as needed
plt.hist(df['Ms_2'], bins=len(df), alpha=0.8, label='Ms_2')  # Adjust the number of bins as needed
plt.title('Histogram of Ms_1 and Ms_2')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Histograms plotted by groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Thus, we can create a histograms of one quantitative variable grouped by another categorical variables.

```python
# Create histograms of the "Ms_1" grouped by "Type"

for type_name, group_data in df.groupby('Type'):
    plt.hist(group_data['Ms_1'], bins=len(df), alpha=0.5, label=type_name)

# Add labels and legend
plt.title('Histogram of Ms_1 Grouped by Type')
plt.xlabel('Ms_1 Values')
plt.ylabel('Frequency')
plt.legend(title='Type')
plt.show()
```

# Boxplots

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. Boxplots and Histograms often complement each other and help an analyst get more information about the data

```python
# Create the boxplot of the "Ms_1" amounts
plt.boxplot(df['Ms_1'])
plt.title('Boxplot of Ms_1')
plt.ylabel('Ms_1 Values')
plt.show()
```

```python
# Create the boxplot of the "tips" amounts

plt.boxplot(df['Ms_2'])
plt.title('Boxplot of Ms_2')
plt.ylabel('Ms_2 Values')
plt.show()
```

```python
# Create the boxplots of the "Ms_1" and of the "Ms_3" amounts

plt.boxplot([df['Ms_1'], df['Ms_3']])
plt.title('Boxplots of Ms_1 and Ms_3')
plt.ylabel('Values')
plt.xticks([1, 2], ['Ms_1', 'Ms_3'])  # Assign labels to x-axis ticks
plt.show()
```

# Boxplots plotted by groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Thus, we can create a side-by-side boxplots of one quantitative variable grouped by another categorical variables.

```python
# Create side-by-side boxplots of the "Ms_1" grouped by "Type"

grouped_data = [group['Ms_1'] for name, group in df.groupby('Type')]

# Create side-by-side boxplots
plt.boxplot(grouped_data, labels=df['Type'].unique())
plt.title('Boxplots of Ms_1 Grouped by Type')
plt.xlabel('Type')
plt.ylabel('Ms_1 Values')
plt.show()
```

# Histograms and boxplots plotted by groups

We cal also create both boxplots and histograms of one quantitative variable grouped by another categorical variables


```python
# Create a boxplot and histogram of the "Ms_1" grouped by "Type"
# Create subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Group the data by 'Type'
grouped_data = [group['Ms_1'] for name, group in df.groupby('Type')]

# Create boxplot
axs[0].boxplot(grouped_data, labels=df['Type'].unique())
axs[0].set_title('Boxplot of Ms_1 Grouped by Type')
axs[0].set_xlabel('Type')
axs[0].set_ylabel('Ms_1 Values')

# Create histogram
for name, group in df.groupby('Type'):
    axs[1].hist(group['Ms_1'], bins=len(df), alpha=0.5, label=name)
axs[1].set_title('Histogram of Ms_1 Grouped by Type')
axs[1].set_xlabel('Ms_1 Values')
axs[1].set_ylabel('Frequency')
axs[1].legend()

plt.tight_layout()
plt.show()
```

# Scatter plot

Plot values of one variable versus another variable to see how they are correlated

```python
# scatter plot between two variables

plt.scatter(df['Ms_1'], df['Ms_2'])
plt.title('Scatter Plot between Ms_1 and Ms_2')
plt.xlabel('Ms_1')
plt.ylabel('Ms_2')
plt.show()
```

```python
# scatter plot between two variables (one categorical)
df["Type encoded"] = df.Type.replace({"Iris-setosa": 1,"Iris-versicolor": 2, "Iris-virginica": 3})

plt.scatter(df['Ms_1'], df['Type encoded'])
plt.title('Scatter Plot between Ms_1 and Type')
plt.xlabel('Ms_1')
plt.ylabel('Type (Encoded)')
plt.yticks(df['Type encoded'].unique(), df['Type'].unique())
plt.show()
```

```python
# scatter plot between two variables (one categorical)
df["Type encoded"] = df.Type.replace({"Iris-setosa": 1,"Iris-versicolor": 2, "Iris-virginica": 3})

plt.scatter(df['Ms_2'], df['Type encoded'])
plt.title('Scatter Plot between Ms_2 and Type')
plt.xlabel('Ms_2')
plt.ylabel('Type (Encoded)')
plt.yticks(df['Type encoded'].unique(), df['Type'].unique())
plt.show()
```

```python
# scatter plot between two variables grouped according to a categorical variable

grouped_data = df.groupby('Type')


for name, group in grouped_data:
    plt.scatter(group['Ms_1'], group['Ms_2'], label=name)


plt.title('Scatter Plot between Ms_1 and Ms_2 Grouped by Type')
plt.xlabel('Ms_1')
plt.ylabel('Ms_2')
plt.legend(title='Type')
plt.show()
```

```python
# scatter plot between two variables grouped according to a categorical variable and with size of markers


grouped_data = df.groupby('Type')


for name, group in grouped_data:
    plt.scatter(group['Ms_1'], group['Ms_2'], s=group['Ms_3']*50, alpha=0.8, label=name)


plt.title('Scatter Plot between Ms_1 and Ms_2 Grouped by Type with Marker Size representing Ms_3')
plt.xlabel('Ms_1')
plt.ylabel('Ms_2')
plt.legend(title='Type')
plt.show()
```
