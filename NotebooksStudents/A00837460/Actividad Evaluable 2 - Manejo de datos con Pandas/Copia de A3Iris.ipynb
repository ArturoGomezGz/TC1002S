2. Plot the histograms for each of the quantitative variables

```python
sb.histplot(df.Sepal_Length)
plt.show()
sb.histplot(df.Sepal_Width)
plt.show()
sb.histplot(df.Petal_Length)
plt.show()
sb.histplot(df.Petal_Width)
plt.show()
sb.histplot(df.Flower_Name)
plt.show()
```

3. Plot the boxplots for each of the quantitative variables

```python
sb.boxplot(df.Sepal_Length)
plt.title('Sepal_Length')
plt.show()
sb.boxplot(df.Sepal_Width)
plt.title('Sepal_Width')
plt.show()
sb.boxplot(df.Petal_Length)
plt.title('Petal_Length')
plt.show()
sb.boxplot(df.Petal_Width)
plt.title('petal_Width')
plt.show()
sb.boxplot(df.Flower_Name)
plt.title('Flower_Name')
plt.show()
```

4. Plot the boxplots of the petal width grouped by type of flower

```python
sb.boxplot(x='Petal_Width', y='Flower_Name', data=df, hue='Flower_Name')
plt.show()
```

5. Plot the boxplots of the sepal length grouped by type of flower

```python
sb.boxplot(x='Sepal_Length', y='Flower_Name', data=df, hue='Flower_Name')
plt.show()
```
