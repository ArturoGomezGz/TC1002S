## A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
#Number of initial clusters: 7
kmeans = KMeans(n_clusters=7, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# Get the centroids
centroids = kmeans.cluster_centers_

# Plotting the clusters for the first two features (x1 and x2)
plt.figure(figsize=(8, 6))
plt.scatter(df['x1'], df['x2'], c=df['Cluster'], cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x', label='Centroids')
plt.title('KMeans Clustering (7 clusters) based on all features')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```

## A5 Elbow plot

Compute the Elbow plot

```python
cluster_range = range(1, 11)
inertia = []  

# Compute KMeans for each number of clusters and store the inertia
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)  
    inertia.append(kmeans.inertia_)  

# Plot the elbow plot
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

```python
#K-Means with 3
newDf = df.drop(['Cluster'], axis=1)
means = KMeans(n_clusters=3, random_state=42)
newDf['Cluster'] = means.fit_predict(newDf)

# Get the centroids
centroids = means.cluster_centers_

# Plotting the clusters for the first two features (x1 and x2)
plt.figure(figsize=(8, 6))
plt.scatter(df['x1'], df['x2'], c=newDf['Cluster'], cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x', label='Centroids')
plt.title('KMeans Clustering (3 clusters) based on all features')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='Cluster')
plt.legend()
plt.show()
```
