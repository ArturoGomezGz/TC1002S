```python
# url string that hosts our .csv file
url = Ruta + "datasets/iris/iris.csv"

# Read the .csv file and store it as a pandas Data Frame
df = pd.read_csv(url)
df
```

```python
df.describe()
```

```python
# Check for missing data and column names
missing_data = df.isnull().sum()
column_names = df.columns

print("Missing Data:")
print(missing_data)

print("\nColumn Names:")
print(column_names)
```

```python
# Add column names
column_names = ["Longitud del sépalo", "Ancho del sépalo", "Longitud del pétalo", "Ancho del pétalo", "Especies (Tipo de flor)"]
df.columns = column_names

# Print DataFrame with column names
df
```

```python
# Create a new dataset containing only petal width, petal length, and the type of flower
petal_data = df[["Longitud del pétalo", "Ancho del pétalo", "Especies (Tipo de flor)"]]

# Create a new dataset containing only sepal width, sepal length, and the type of flower
sepal_data = df[["Longitud del sépalo", "Ancho del sépalo", "Especies (Tipo de flor)"]]

# Print the datasets
print("Petal Data:")
print(petal_data)

print("\nSepal Data:")
print(sepal_data)
```

```python
# Define a mapping dictionary for encoding flower types
flower_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}

# Create a new column in the original DataFrame with encoded flower types
df['Encoded_Flower_Type'] = df['Especies (Tipo de flor)'].map(flower_mapping)

# Create a new dataset containing sepal width, sepal length, and the encoded flower type
encoded_sepal_data = df[["Longitud del sépalo", "Ancho del sépalo", "Encoded_Flower_Type"]]

# Print the new dataset
print("Encoded Sepal Data:")
print(encoded_sepal_data)
```
