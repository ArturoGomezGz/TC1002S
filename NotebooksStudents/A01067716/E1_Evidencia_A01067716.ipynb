## A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

k = 3
km = KMeans(n_clusters=k, n_init="auto", random_state=42)
typePredict = km.fit_predict(df[['x1', 'x2']])

df['Cluster'] = typePredict

# resultados
plt.figure(figsize=(8, 6))
plt.scatter(df['x1'], df['x2'], c=df['Cluster'], cmap='viridis', s=50)
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroides')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title(f'K-means Clustering con {k} Clusters')
plt.legend()
plt.show()
```

Add to your dataset a column with the estimated cluster to each data point

```python
from sklearn.cluster import KMeans

k = 3
km = KMeans(n_clusters=k, n_init="auto", random_state=42)
typePredict = km.fit_predict(df[['x1', 'x2']])

# Añadir una columna al dataset
df['Cluster_Estimado'] = typePredict

df.head()
```

## A5 Elbow plot

Compute the Elbow plot

```python

k_values = range(1, 11)
sse = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init="auto", random_state=42)
    kmeans.fit(df[['x1', 'x2']])
    sse.append(kmeans.inertia_)

# Graficar
plt.plot(k_values, sse, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Sum of Squared Error (SSE)')
plt.title('Elbow Method For Optimal k')
plt.show()
```
