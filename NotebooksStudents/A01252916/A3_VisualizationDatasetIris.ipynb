# Descrirpion of the quantitative variables
Understanding the characteristics of our analyzed data is essential for informed analysis. Through the use of histogram plots for each of our quantitative variables, we can glean valuable insights into the distribution of our data. Histograms provide a visual representation of the frequency of values within a variable, enabling us to discern central tendencies, variability, and comparisons with other quantitative variables.

On the other hand, box plots also play a crucial role in visualizing the distributional characteristics of our data, particularly through quartiles. Combining these two plot techniques creates a robust analytical space for comprehending our data.

* Sepal Length:
Histogram analysis reveals a central tendency of 6 cm for sepal length. Upon examining sepal length across different flower types through box plots, it becomes evident that Iris-setosa exhibits the smallest sepal length, followed by Iris-versicolor, and finally Iris-virginica with the largest sepal length.

* Sepal Width:
Histogram information indicates a clear tendency towards a central distribution of 3.0 cm for sepal width. Additionally, it is noteworthy that sepal width tends to be half the size of sepal length, with the majority of flowers presenting a width of 3.0 cm.

* Petal Length:
Analysis of petal length reveals a mean of 3.7 cm, with some variation observed in the central tendency. The majority of values oscillate between 1-2 cm and 4.5-5 cm, which is reasonable given the mean of 3.7 cm across different flower types.

* Petal Width:
Histogram analysis shows a distribution ranging between 0-0.3 cm and 1.25-2 cm, indicating a wide variety of widths across petals. Additionally, there is a discernible difference between Iris-setosa, which has approximately half the width of Iris-versicolor and one-sixth the width of Iris-virginica.
