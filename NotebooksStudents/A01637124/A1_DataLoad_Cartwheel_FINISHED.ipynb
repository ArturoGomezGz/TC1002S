```python
# Define the col names for the iris dataset
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']

# Dataset url
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Load the dataset
df = pd.read_csv(url, header=None, names=columns)
```

```python
df.head(20)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#How many observations (rows) are in total?
df.shape[0]
#150 rows in total
```

```python
#How many variables (columns) are in total? What do they represent?
df.shape[1]
#5 rows in total
#4 variable represent measures (width and length) of parts of the flower
##2 are measures of the sepal, 2 are measures of the petal
#1 variable represents the type of flower
```

```python
#How many observations are there for each type of flower?
observations_species = df.groupby('Species').size()
print(observations_species)
#50 for each
```
