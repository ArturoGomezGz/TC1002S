## A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
# prompt: Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

from sklearn.cluster import KMeans

# Select the features you want to use for clustering (e.g., the second and third columns)
X = df.iloc[:, [1, 2]]  # Shift to the second and third columns

# Determine the optimal number of clusters (e.g., based on your scatter plot)
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the KMeans model to your data
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Add the cluster labels to your DataFrame
df['cluster'] = labels

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(df.iloc[:, 1], df.iloc[:, 2], c=df['cluster'], cmap='viridis')  # Shift to second and third columns
plt.xlabel('X1')  # Replace with actual name if needed
plt.ylabel('X2')  # Replace with actual name if needed
plt.title('KMeans Clustering')
plt.show()

```

Add to your dataset a column with the estimated cluster to each data point

```python
# prompt: Add to your dataset a column with the estimated cluster to each data point

# Assuming you already have the 'labels' from your KMeans model

df['cluster'] = labels
```

Print the number associated to each cluster

```python
# prompt: Print the number associated to each cluster

for cluster_num in df['cluster'].unique():
  print(f"Cluster {cluster_num}:")
  print(df[df['cluster'] == cluster_num].shape[0])
```

Print the centroids

```python
# prompt: Print the centroids

print(kmeans.cluster_centers_)
```

Plot a scatter plot of your data using different color for each cluster. Also plot the centroids

```python
# prompt: Plot a scatter plot of your data using different color for each cluster. Also plot the centroids

# Plot a scatter plot of your data using different color for each cluster. Also plot the centroids
plt.figure(figsize=(8, 6))

# Plot the data points, using the second and third columns and color by cluster
plt.scatter(df.iloc[:, 1], df.iloc[:, 2], c=df['cluster'], cmap='viridis')

# Plot the centroids, which correspond to the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')

# Adjust the labels to reflect the correct columns
plt.xlabel('X1')  # Replace with actual column name if needed
plt.ylabel('X2')  # Replace with actual column name if needed
plt.title('KMeans Clustering with Centroids')
plt.legend()
plt.show()

```

## A5 Elbow plot

Compute the Elbow plot

```python
# prompt: Compute the Elbow plot

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```
