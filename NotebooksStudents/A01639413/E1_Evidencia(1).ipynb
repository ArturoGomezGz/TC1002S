## A4 Kmeans

Do Kmeans clustering assuming a number of clusters accorging to your scatter plot

```python
# Import the StandardScaler class
from sklearn.preprocessing import StandardScaler
# Import the KMeans class
from sklearn.cluster import KMeans
# Drop any non-numerical columns like 'Unnamed: 0'
data = df.drop(columns=['Unnamed: 0'])

# Standardize the data before clustering
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# 1. KMeans clustering with 3 clusters (you can change this based on your analysis)
kmeans = KMeans(n_clusters=6, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

```

Add to your dataset a column with the estimated cluster to each data point

```python

# 2. Add a new column with the estimated cluster for each data point
df['Cluster'] = clusters
df

```

Print the centroids

```python
centroids = kmeans.cluster_centers_
print("\nCentroids of the clusters:")
print(centroids)
```

Print the intertia metric

```python
# 5. Print the inertia metric
inertia = kmeans.inertia_
print("\nInertia (Sum of squared distances to centroids):")
print(inertia)
```

Plot a scatter plot of your data using different color for each cluster. Also plot the centroids

```python
# 6. Scatter plot of the data with different colors for each cluster
plt.figure(figsize=(8, 6))

# Plot the scatter plot with cluster colors
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters, cmap='viridis', alpha=0.6, label="Data points")

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label="Centroids")

plt.title('KMeans Clustering with 6 Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

## A5 Elbow plot

Compute the Elbow plot

```python
data = df.drop(columns=['Unnamed: 0'])

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# 1. Compute the inertia for different number of clusters
inertia_values = []
cluster_range = range(1, 11)  # We'll test for 1 to 10 clusters

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia_values.append(kmeans.inertia_)

# 2. Plot the Elbow plot
plt.figure(figsize=(8, 6))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='--', color='b')
plt.title('Elbow Plot for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia (Sum of Squared Distances)')
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```
