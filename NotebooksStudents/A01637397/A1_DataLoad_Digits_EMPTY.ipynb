# Digits data set v1

- Data from scikit-learn https://scikit-learn.org/stable/auto_examples/datasets/plot_digits_last_image.html

- More details in https://archive.ics.uci.edu/ml/datasets/Pen-Based+Recognition+of+Handwritten+Digits


```python
# Import the dataset
from sklearn.datasets import load_digits

# Load the dataset
DS = load_digits()
```

```python
# Print keys
print(DS.keys())
```

```python
# Explore the dataset

# Data as vector
data_vector = DS['data']
print("Data as vector:\n", data_vector)

# Target/Class/Group
target = DS['target']
print("\nTarget/Class/Group:\n", target)

# Data as matrix
data_matrix = DS['images']
print("\nData as matrix:\n", data_matrix)

# Name of the variables
print("\nName of the variables:\n", DS.keys())

# Name of the classes/groups
target_names = DS['target_names']
print("\nName of the classes/groups:\n", target_names)
```

```python
# Select one of the observations
i = 345
```

```python
# Show the data vector of the selected observation
selected_data_vector = data_vector[i]
print("Data vector of the selected observation:\n", selected_data_vector)
```

```python
# Show the data matrix of the selected observation
selected_data_matrix = data_matrix[i]
print("Data matrix of the selected observation:\n", selected_data_matrix)
```

```python
# Show the number of the selected observation
print("Number of the selected observation:", target[i])
```

```python
# Display one of the observations (zeros as white)
import matplotlib.pyplot as plt
plt.imshow(selected_data_matrix, cmap='gray_r')
plt.title(f"Digit: {target[i]}")
plt.show()
```

```python
# Display one of the observations (zeros as black)
plt.imshow(selected_data_matrix, cmap='gray')
plt.title(f"Digit: {target[i]}")
plt.show()
```

```python
# Display 4 observations in a row (zeros as white)
plt.figure(figsize=(12, 3))
for index in range(4):
    plt.subplot(1, 4, index + 1)
    plt.imshow(data_matrix[index], cmap='gray_r')
    plt.title(f"Digit: {target[index]}")
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Display 4 observations in a row (zeros as black)
plt.figure(figsize=(12, 3))
for index in range(4):
    plt.subplot(1, 4, index + 1)
    plt.imshow(data_matrix[index], cmap='gray')
    plt.title(f"Digit: {target[index]}")
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
# Convert to a pandas dataframe
import pandas as pd

# Create the dataframe
df = pd.DataFrame(data_vector, columns=DS['feature_names'])
df['target'] = target

# Print the dataset
print(df)
```

# Digits data set v2

- Data from scikit-learn https://scikit-learn.org/stable/auto_examples/datasets/plot_digits_last_image.html

- More details in https://archive.ics.uci.edu/ml/datasets/Pen-Based+Recognition+of+Handwritten+Digits

```python
# Import the dataset
from sklearn.datasets import load_digits

# Load the dataset
X, y = load_digits(return_X_y=True)
```

```python
# Understand X (type and shape)
print("Type of X:", type(X))
print("Shape of X:", X.shape)
```

```python
# Understand y (type and shape)
print("Type of y:", type(y))
print("Shape of y:", y.shape)
```

```python
# -------------------------------
# Select some of the images
iimag = [  0,   11,   28,  42,    56,  70,    84,
          91,  115,  280, 420,   560, 700,   906,
         929,  970,  999, 1003, 1077, 1104, 1205,
        1206, 1153, 1205, 1257, 1378, 1407, 1509]

# -------------------------------
# Plot selected images (zeros as black)
plt.figure(figsize=(12,8))
for i, i2p in enumerate(iimag):

    Xi = X[i2p,:]
    yi = y[i2p]

    plt.subplot(4,7,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(Xi.reshape(8,8), cmap=plt.cm.gray)
    plt.title("Label: {0:.0f}".format(yi))

    # Verify data
    #minX = np.min(Xi)
    #maxX = np.max(Xi)
    #print("min max")
    #print(minX, maxX)
    #print(" ")

plt.show()
```
